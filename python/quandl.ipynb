```python
import quandl as q
q.ApiConfig.api_key = API_KEY
d = q.get('BCHAIN/MKPRU')
d['SMA'] = d['Value'].rolling(100).mean()
d.loc['2016-1-1':].plot(title='BTC/USD exchange rate', figsize=(10, 6))
```

```python
cs = q.get('UMICH/SOC22', start_date='2008-1-1', end_date='2021-4-1')
ax = plt.gca()
columns = cs.columns[0:6]
cs.plot(ax=ax, y=columns, ylabel='%', title='income distribution', figsize=(10, 8), kind='area')
```
